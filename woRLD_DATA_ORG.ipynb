```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

path = "elec.csv"
df = pd.read_csv(path)
df.info()
df.head(20)
```

```python

df.drop(['Series Name','Series Code','Country Code','2021 [YR2021]'], axis = 1, inplace = True)
df.rename(columns={'1990 [YR1990]': "1990"}, inplace=True)
df.rename(columns={'2000 [YR2000]': "2000"}, inplace=True)
df.rename(columns={'2012 [YR2012]': "2012"}, inplace=True)
df.rename(columns={'2013 [YR2013]': "2013"}, inplace=True)
df.rename(columns={'2014 [YR2014]': "2014"}, inplace=True)
df.rename(columns={'2015 [YR2015]': "2015"}, inplace=True)
df.rename(columns={'2016 [YR2016]': "2016"}, inplace=True)
df.rename(columns={'2017 [YR2017]': "2017"}, inplace=True)
df.rename(columns={'2018 [YR2018]': "2018"}, inplace=True)
df.rename(columns={'2019 [YR2019]': "2019"}, inplace=True)
df.rename(columns={'2020 [YR2020]': "2020"}, inplace=True)
df.head()
```

```python
world = df.groupby("Country Name")["1990","2000","2012","2013","2014","2015","2016","2017","2018","2019","2020"].sum().reset_index()
world.head()
```

```python
### Find top 20 countries with maximum number of confirmed cases
from matplotlib import container
top_5 = world.sort_values(by='2012', ascending=False).head(5)
top_5.head()
```

```python
plt.figure(figsize=(12,10))
plot = sns.barplot(x=top_5['2012'], y=top_5['Country Name'])
plot.bar_label(plot.containers[0])

plt.show()
```
